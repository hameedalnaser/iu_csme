# Feature Engineering

```python
# extracting the historical weather data of berlin city from https://open-meteo.com/en/docs/historical-weather-api
#downloading the data as CSV that match the sickness table dates

weatherdata = pd.read_csv("C:/Users/MSI-PC/case_study/data/weather.csv")
weatherdata.head()
```

```python

#extracting featuered data from date such as month, weeks, weekends etc.
sickdata['month']=pd.to_datetime(sickdata['date']).dt.month
sickdata['yearweek']=pd.to_datetime(sickdata['date']).dt.week
sickdata['monthday']=pd.to_datetime(sickdata['date']).dt.day
sickdata['weekday']=pd.to_datetime(sickdata['date']).dt.weekday
sickdata['dayofyear']=pd.to_datetime(sickdata['date']).dt.dayofyear
sickdata['quarter']=pd.to_datetime(sickdata['date']).dt.quarter
sickdata["weekend"]=sickdata['weekday'].replace([0,1,2,3,4,5,6],[0,0,0,0,0,1,1])
sickdata.head()
```

```python
#merging the featuered sickness data and weather data together
featuereddata = pd.DataFrame()
featuereddata['n_sick']=sickdata['n_sick']
featuereddata['calls']=sickdata['calls']
featuereddata['n_duty']=sickdata['n_duty']
featuereddata['sby_need']=sickdata['sby_need']
featuereddata['month']=sickdata['month']
featuereddata['yearweek']=sickdata['yearweek']
featuereddata['weekday']=sickdata['weekday']
featuereddata['weekend']=sickdata['weekend']
featuereddata['temperature_2m_max']=weatherdata['temperature_2m_max (°C)']
featuereddata['temperature_2m_min']=weatherdata['temperature_2m_min (°C)']
featuereddata['rain_sum']=weatherdata['rain_sum (mm)']
featuereddata['snowfall_sum']=weatherdata['snowfall_sum (cm)']
featuereddata['precipitation_hours']=weatherdata['precipitation_hours (h)']
featuereddata['windspeed_10m_max']=weatherdata['windspeed_10m_max (km/h)']
featuereddata['dayofyear']=sickdata['dayofyear']
featuereddata['quarter']=sickdata['quarter']
featuereddata.head()
```

# EDA Exploratory Data Analysis

The final target is to find the sby_need values, and this value is logically related to the number of calls. more details will be discovered in this section

```python
#discovering the relationship between the number of calls and the sby_need
plt.scatter(featuereddata['calls'],featuereddata['sby_need'])
plt.xlabel('number of sby_need')
plt.ylabel('number of calls');
plt.savefig('sby_need_vs_calls')
```

from the scatter plot, it seems that the relationship between the number of calls and the sby_need is exponential regression.

To find the number of calls we need to explore the relationship between it and the other attributes.

```python
# exploring the number of standby drivers 
sickdata=sickdata.set_index('date')
sickdata['sby_need'].plot(style='.',ylabel='sby_need', figsize=(15,5), title='the number of standby needs drivers')
plt.savefig('sby_need_vs_date')
```

```python
# exploring the number of calls
sickdata['calls'].plot(style='.',ylabel='sby_need', figsize=(15,5), title='the number of calls')
plt.savefig('calls_vs_date')
```

```python
#calculating the correlation coefficient
corrmat = featuereddata.corr()
cmap = sns.diverging_palette(260,-10,s=50, l=75, n=6, as_cmap=True)
plt.subplots(figsize=(18,18))
sns.heatmap(corrmat,cmap= cmap,annot=True, square=True)
plt.savefig('correlation cofficient')
```

The most strong relationships are found between calls and maximum temperature, minimum temperature, weekday, weekend, dayofyear, yearweek and snowfall sum.

```python
#here are several EDA techniques to understand more about the calls parameter
fig, ax=plt.subplots(figsize=(15,8))
sns.boxplot(data=featuereddata, x='quarter', y='calls')
plt.savefig('Box_plot_calls_in_each_quarter')
```

```python
fig, ax=plt.subplots(figsize=(15,8))
sns.boxplot(data=featuereddata, x='month', y='calls')
plt.savefig('Box_plot_calls_in_each_month')
```

```python
fig, ax=plt.subplots(figsize=(15,8))
sns.boxplot(data=featuereddata, x='weekday', y='calls')
plt.savefig('Box_plot_calls_in_each_week')
```
